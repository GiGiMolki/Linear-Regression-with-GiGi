# 04 — Simple Linear Regression using Scikit-Learn

In this notebook, we’ll:
- Use sklearn’s `LinearRegression` class
- Train a model using synthetic data
- Extract slope and intercept
- Visualize the fit
- Evaluate with R² Score

## Step 1: Import Libraries

We'll use:
- NumPy for numerical arrays
- Matplotlib + Seaborn for plots
- Scikit-learn’s LinearRegression to build the model

```python
# ✅ Step 1: Import Libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (6, 3)
```

## Step 2: Generate Data

We simulate a dataset with the equation `y = 3x + 4 + noise`.

```python
# ✅ Step 2: Generate Synthetic Data
np.random.seed(42)
X = np.linspace(0, 10, 50)
noise = np.random.normal(0, 2, X.shape)
y = 3 * X + 4 + noise
```

## Step 3: Reshape Data

We reshape X and y to the proper format for `sklearn`, where:
- `X` must be a 2D array: shape `(n_samples, n_features)`
- `y` must be a 1D or 2D array: shape `(n_samples,)` or `(n_samples, 1)`

```python
# ✅ Step 3: Reshape Data
X = X.reshape(-1, 1)
y = y.reshape(-1, 1)
```

## Step 4: Train Model

We instantiate the `LinearRegression()` model and fit it using `.fit(X, y)`.

```python
# ✅ Step 4: Create and Train the Model
model = LinearRegression()
model.fit(X, y)
```

## Step 5: Model Parameters

After training, we extract:
- `intercept_` (θ₀)
- `coef_` (θ₁)

The final model is:

\[
\hat{y} = {slope:.2f}x + {intercept:.2f}
\]

```python
# ✅ Step 5: Access Model Parameters
intercept = model.intercept_[0]
slope = model.coef_[0][0]

print(f"Intercept: {intercept:.2f}")
print(f"Slope: {slope:.2f}")
```

## Step 6: Make Predictions

We use `.predict(X)` to calculate predicted values.

```python
# ✅ Step 6: Predict
y_pred = model.predict(X)
```

## Step 7: Visualize Fit

This shows the regression line learned by the model overlaid on the original data.

```python
# ✅ Step 7: Plot the Fit
plt.scatter(X, y, label="Actual Data")
plt.plot(X, y_pred, color="red", label="Prediction (Best Fit Line)")
plt.title(f"Regression Line: y = {slope:.2f}x + {intercept:.2f}")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```

## Step 8: R² Score

R² score (Coefficient of Determination) measures how well the regression line fits the data.
- R² = 1: Perfect fit
- R² ≈ 0: Poor fit

```python
# ✅ Step 8: Evaluate Model
r2 = r2_score(y, y_pred)
print(f"R² Score: {r2:.4f}")
```
