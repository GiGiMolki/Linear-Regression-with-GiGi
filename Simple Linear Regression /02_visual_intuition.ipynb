## Step 2: Visualize Data Distribution

We simulate data where:
- `y` is linearly dependent on `x` via `y = 3x + 4`
- We add random Gaussian noise to make it realistic

This gives us a scatter plot that looks like a cloud around a straight line.

```python
# âœ… Step 2: Create Synthetic Linear Data
np.random.seed(42)

X = np.linspace(0, 10, 50)                   # 50 evenly spaced points between 0 and 10
noise = np.random.normal(0, 2, X.shape)      # Add Gaussian noise
y = 3 * X + 4 + noise                        # True relation: y = 3x + 4 + noise

# Scatter plot
plt.scatter(X, y, color='blue', label='Data Points')
plt.title('Scatter Plot of Synthetic Data')
plt.xlabel('X (Independent Variable)')
plt.ylabel('y (Dependent Variable)')
plt.legend()
plt.show()
```

## Step 3: Plot the True Regression Line

We overlay the **true underlying regression line** (`y = 3x + 4`) on the scatter plot.

Visually, this line appears to best capture the central trend in the noisy data.

```python
# âœ… Step 3: Plot the Ideal Regression Line
m = 3
c = 4
y_pred = m * X + c

plt.scatter(X, y, label='Data Points', alpha=0.7)
plt.plot(X, y_pred, color='red', label='Regression Line (y = 3x + 4)')
plt.title('Regression Line Over Data')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

## Step 4: Show Residuals

Residuals are the **vertical errors** between each true `y` value and the predicted `Å·`.

We draw gray dashed lines from each point to the regression line to visualize these errors. The regression line is optimized to minimize the **sum of squared residuals**.

```python
# âœ… Step 4: Visualize Residuals
for i in range(len(X)):
    plt.plot([X[i], X[i]], [y[i], y_pred[i]], color='gray', linestyle='--')

plt.scatter(X, y, label='Actual Points', color='blue')
plt.plot(X, y_pred, color='red', label='Regression Line')
plt.title('Residuals (Vertical Distance from Points to Line)')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

## Step 5: Try Various Lines

We try out different combinations of slopes (`m`) and intercepts (`c`) to see how they fit the data.

Some lines underfit, some overfit, but only one minimizes the **total squared error** â€” that's the optimal one we aim to find using the **least squares method**.

```python
# âœ… Step 5: Try Different Lines
slopes = [1, 2, 3, 4, 5]
intercepts = [2, 4, 6]

plt.scatter(X, y, color='black', label='Data')

for m in slopes:
    for c in intercepts:
        y_temp = m * X + c
        plt.plot(X, y_temp, alpha=0.4, label=f"m={m}, c={c}")

plt.title("Multiple Candidate Regression Lines")
plt.xlabel("X")
plt.ylabel("y")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

## Step 6: Good Fit vs Bad Fit

Finally, we compare a poorly chosen line (wrong slope & intercept) vs the optimal regression line.

The **green line** (y = 3x + 4) fits well and minimizes the residuals.

The **gray dashed line** (y = x + 10) is clearly off â€” both visually and mathematically.

```python
# âœ… Step 6: Compare Good Fit vs Bad Fit
bad_fit = 1 * X + 10
good_fit = 3 * X + 4

plt.scatter(X, y, label='Data', color='black')
plt.plot(X, bad_fit, color='gray', linestyle='--', label='Bad Fit: y = x + 10')
plt.plot(X, good_fit, color='green', label='Good Fit: y = 3x + 4')
plt.title("Good Fit vs Bad Fit")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```
