# 06 — Simple Linear Regression using Gradient Descent (from scratch)

In this notebook, you will:
- Understand the logic of gradient descent
- Implement cost and gradient functions manually
- Update parameters iteratively
- Visualize convergence and final model

## Step 1: Libraries

We'll use NumPy for math and Matplotlib + Seaborn for visualization.

```python
# ✅ Step 1: Import Libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (6, 3)
```

## Step 2: Data Generation

Same data as before:
- theta_1 = 3, theta_0 = 4
- Added Gaussian noise to simulate real-world data

```python
# ✅ Step 2: Generate Synthetic Data
np.random.seed(42)
X = np.linspace(0, 10, 50)
noise = np.random.normal(0, 2, X.shape)
y = 3 * X + 4 + noise
```

## Step 3 : Reshape X and y

```python
# ✅ Step 3: Reshape X and y
X = X.reshape(-1, 1)
y = y.reshape(-1, 1)
```

## Step 4: Initialization

We initialize:
- theta_0, theta_1 = 0
- Learning rate alpha = 0.01
- Iterations = 1000

```python
# ✅ Step 4: Initialize Parameters and Settings
theta0 = 0
theta1 = 0
alpha = 0.01  # Learning rate
n_iterations = 1000
n = len(X)
```

## Step 5: Gradient Descent Algorithm

We loop:
- Predict hat{y} = theta_0 + theta_1 * x
- Compute error and cost (MSE)
- Update θ₀ and θ₁ using gradients

```python
# ✅ Step 5: Gradient Descent Loop
cost_history = []

for i in range(n_iterations):
    y_pred = theta0 + theta1 * X
    error = y_pred - y

    cost = (1/n) * np.sum(error**2)
    cost_history.append(cost)

    # Gradients
    d_theta0 = (2/n) * np.sum(error)
    d_theta1 = (2/n) * np.sum(error * X)

    # Update parameters
    theta0 -= alpha * d_theta0
    theta1 -= alpha * d_theta1
```

## Step 6: Final Parameters


```python
# ✅ Step 6: Final Parameters
print(f"θ₀ (intercept): {theta0:.4f}")
print(f"θ₁ (slope):     {theta1:.4f}")
```

## Step 7: Cost Convergence

This shows how the loss function (MSE) decreased over time as the model learned.

```python
# ✅ Step 7: Visualize Cost Convergence
plt.plot(range(n_iterations), cost_history, color='purple')
plt.title("Cost Function Convergence")
plt.xlabel("Iterations")
plt.ylabel("MSE Loss")
plt.show()
```

This line chart visualizes how the MSE decreases over each iteration.
- Left of the plot: High cost at the beginning → poor model
- Right of the plot: Cost levels off → model is approaching the best fit
- The line becomes flat → this means the model converged

## Step 8: Final Line Visualization

You can now visually confirm that the learned line fits the data well.

```python
# ✅ Step 8: Plot Final Regression Line
y_final = theta0 + theta1 * X

plt.scatter(X, y, label="Actual Data")
plt.plot(X, y_final, color="red", label=f"Fit: y = {theta1:.2f}x + {theta0:.2f}")
plt.title("Linear Regression Fit (Gradient Descent)")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```
